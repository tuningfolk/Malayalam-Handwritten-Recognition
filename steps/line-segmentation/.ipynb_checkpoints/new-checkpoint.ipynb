```python
 def pre_process_img(img):
        gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        # blur image
        smoothed_img = cv2.blur(gray_img, (3, 3), anchor=(-1, -1))

        _, thresh = cv2.threshold(smoothed_img, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
        return thresh, gray_img

        # cv2.imwrite(self.output_path + "/Binary_image.jpg", self.thresh)
```

```python
def mergeRects(rects):
    """
     rects = [(x,y,w,h),(x,y,w,h),...]
     return merged rects
    """
    merged_rects = []
    for i in range(len(rects)):
        is_repeated = False
        for j in range(i+1, len(rects)):
            rect_tmp = intersection(rects[i], rects[j])
            # Check for intersection / union
            if rect_tmp == ():
                continue
            rect_tmp_area = rect_tmp[2] * rect_tmp[3]
            rect_i_area = rects[i][2] * rects[i][3]
            rect_j_area = rects[j][2] * rects[j][3]

            if ((rect_tmp_area == rect_i_area) or (rect_tmp_area == rect_j_area)):
                is_repeated = True
                rect_tmp = union(rects[i], rects[j]) # Merging
                merged_rect = rect_tmp
                # Push in merged rectangle after checking all the inner loop.
                if j == len(rects) - 2:
                    merged_rects.append(merged_rect)
                # Update the current vector.
                rects[j] = merged_rect

        if is_repeated == False:
            merged_rects.append(rects[i])

    return merged_rects
```

```python
def find_contours():
        thresh_clone = thresh_img.copy()
    
        contours = 0
        contours,hierachy = cv2.findContours(image=thresh_clone, mode=cv2.RETR_LIST, method=cv2.CHAIN_APPROX_NONE, offset=(0,0))

        bounding_rect = []
        for c in contours:
            cv2.approxPolyDP(c, 1, True, c)  # apply approximation to polygons
            bounding_rect.append(cv2.boundingRect(c))
        del (bounding_rect[-1])  # coz the last one is a contour containing whole image
        img_clone = img.copy()
        # check intersection among rects
        merged_rect = mergeRects(bounding_rect)

        # draw rects on image
        for r in merged_rect:
            (x, y, w, h) = r
            cv2.rectangle(img_clone, (x, y), (x + w, y + h), (0, 0, 255), 2, 8, 0)
        # cv2.imwrite(output_path + "/contours.jpg", img_clone)
        contours = merged_rect  # save contours AS rectangles
        # # contours = []
        # # for rect in merged_rect:
        # #     (x,y,w,h) = rect
        # #     new_contour = []
        # #     for i in range(x, x + w):
        # #         new_contour.append([i, y])
        # #     for i in range(y+1, y + h - 1):
        # #         new_contour.append([x, i])
        # #         new_contour.append([x + w - 1, i])
        # #     for i in range(x, x + w):
        # #         new_contour.append([i, y+h - 1])
        # #     contours.append(new_contour)
        # print(contours)
        return contours


def union(a,b):
    """
    :param a: rect
    :param b: rect
    :return: union area 
    """
    x = min(a[0], b[0])
    y = min(a[1], b[1])
    w = max(a[0]+a[2], b[0]+b[2]) - x
    h = max(a[1]+a[3], b[1]+b[3]) - y
    return (x, y, w, h)

def intersection(a,b):
    """
    :param a:  rect
    :param b:  rect
    :return: intersection area 
    """
    x = max(a[0], b[0])
    y = max(a[1], b[1])
    w = min(a[0]+a[2], b[0]+b[2]) - x
    h = min(a[1]+a[3], b[1]+b[3]) - y
    if w<0 or h<0: return () # does not intersect
    return (x, y, w, h)
```

```python
CHUNK_NUMBER = 20
CHUNK_TOBE_PROCESSED = 5

def generate_chunks(thresh_img):
        rows, cols = thresh_img.shape[:2]
        width = cols
        chunk_width = int(width / CHUNK_NUMBER)
        chunks = []
        start_pixel = 0
        for i_chunk in range(CHUNK_NUMBER):
            c = Chunk(index=i_chunk, start_col=start_pixel, width=chunk_width,
                      img=thresh_img[0:rows, start_pixel:start_pixel + chunk_width].copy())
            chunks.append(c)

            # cv2.imwrite(self.output_path + "/Chunk" + str(i_chunk) + ".jpg", self.chunks[-1].gray_img)

            start_pixel += chunk_width

        return chunks
```

```python
chunks = generate_chunks(thresh_img)
```

```python
map_valley = {}
predicted_line_height = 0
avg_line_height = 0
```

```python
def connect_valleys( i, current_valley, line, valleys_min_abs_dist):
        """
        i: the chunk's index
        current_valley
        line
        valleys_min_abs_dist:
        """
        if (i <= 0 or len(chunks[i].valleys)==0):
            return line
        # choose the closest valley in right chunk to the start valley
        connected_to = -1
        min_dist = sys.maxsize
        # found valleys's size in this chunk
        valleys_size = len(chunks[i].valleys)
        for j in range(valleys_size):
            valley = chunks[i].valleys[j]
            # Check if the valley is not connected to any other valley
            if valley.used is True:
                continue

            dist = abs(current_valley.position - valley.position)

            if min_dist > dist and dist <= valleys_min_abs_dist:
                min_dist = dist
                connected_to = j

        # Return line if the current valley is not connected any more to a new valley in the current chunk of index i.
        if connected_to == -1:
            return line

        line.valley_ids.append(chunks[i].valleys[connected_to].valley_id)
        v = chunks[i].valleys[connected_to]
        v.used = True
        return connect_valleys(i - 1, v, line, valleys_min_abs_dist)
```

```python
def get_initial_candidate_lines(map_valley):

        height, width = thresh_img.shape[:2]
        chunk_width = int(width / CHUNK_NUMBER)
    
        number_of_heights = 0
        valleys_min_abs_dist = 0
    
        initial_lines = []
    
        # Get the histogram of the first CHUNK_TO_BE_PROCESSED and get the overall average line height.
        for i in range(CHUNK_TOBE_PROCESSED):
            avg_height = chunks[i].find_peaks_valleys(map_valley)

            if avg_height:
                number_of_heights += 1
            valleys_min_abs_dist += avg_height

        valleys_min_abs_dist /= number_of_heights
        print("Estimated avg line height {}".format(valleys_min_abs_dist))
        predicted_line_height = valleys_min_abs_dist

        # Start from the CHUNK_TOBE_PROCESSED chunk
        for i in range(CHUNK_TOBE_PROCESSED - 1, 0, -1):
            if len(chunks[i].valleys) == 0:
                continue

            # Connect each valley with the nearest ones in the left chunks.
            for valley in chunks[i].valleys:
                if valley.used is True:
                    continue

                # Start a new line having the current valley and connect it with others in the left.
                valley.used = True

                new_line = Line(valley.valley_id)
                new_line = connect_valleys(i - 1, valley, new_line, valleys_min_abs_dist)
                new_line.generate_initial_points(chunk_width, img.shape[1], map_valley)

                if len(new_line.valley_ids) > 1:
                    initial_lines.append(new_line)
        print("There are {}".format(len(initial_lines)) + " initial lines")
        return initial_lines, predicted_line_height
```

```python
initial_lines, predicted_line_height = get_initial_candidate_lines(map_valley)
```

```python
def save_image_with_lines(img, initial_lines):
        img_clone = img.copy()

        for line in initial_lines:
            last_row = -1

            for point in line.points:
                img_clone[int(point[0])][int(point[1])] = (0, 0, 255)
                # Check and draw vertical lines if found.
                if last_row != -1 and point[0] != last_row:
                    for i in range(min(int(last_row), int(point[0])), max(int(last_row), int(point[0]))):
                        img_clone[int(i)][int(point[1])] = (0, 0, 255)

                last_row = point[0]

        # cv2.imwrite(img_clone)
        return img_clone
```

```python
def generate_regions():
    if len(initial_lines) == 0:
        return
    # Sort lines by row position
    initial_lines.sort(key=Line().get_min_row_position)
    #reset line regions
    lines_region = []
    avg_line_height = 0
    # Add first region
    r = Region(top=Line(), bottom=initial_lines[0])

    r.update_region(gray_img, 0)
    initial_lines[0].above = r
    lines_region.append(r)

    if r.height < predicted_line_height * 2.5:
        avg_line_height += r.height

    # Add rest of regions.
    for i in range(len(initial_lines)-1):
        top_line = initial_lines[i]
        bottom_line = initial_lines[i + 1]
        # Assign lines to region.
        r = Region(top_line, bottom_line)
        res = r.update_region(gray_img, i)
        # Assign regions to lines
        if top_line.initial_valley_id != -1:
            top_line.below = r

        if bottom_line.initial_valley_id != -1:
            bottom_line.above = r
        if res:
            lines_region.append(r)
            if (r.height < predicted_line_height * 2.5):
                avg_line_height += r.height

    if len(lines_region) > 0:
        avg_line_height /= len(lines_region)
        print("Avg line height is {}".format(int(avg_line_height)))

    return lines_region, avg_line_height
```

```python
lines_region, avg_line_height = generate_regions()
```

```python
primes = []
not_primes_list = [0] * 100007

def sieve():
        not_primes_list[0] = 1
        not_primes_list[1] = 1
        for i in range(2, 100000):
            if not_primes_list[i] == 1:
                continue

            primes.append(i)
            for j in range(i * 2, 100000, i):
                not_primes_list[j] = 1


def addPrimesToList( n, probPrimes):
        for i in range(len(primes)):
            while (n % primes[i]):
                n /= primes[i]
                probPrimes[i] += 1



sieve()
```

```python
def component_belongs_to_above_region( line, contour):
        # Calculate probabilities
        probAbovePrimes = [0] * len(primes)
        probBelowPrimes = [0] * len(primes)
        n = 0

        tl = (contour[0], contour[1])  # top left

        width, height = contour[2], contour[3]

        for i_contour in range(tl[0], tl[0] + width):
            for j_contour in range(tl[1], tl[1] + height):
                if thresh_img[j_contour][i_contour] == 255:
                    continue

                n += 1

                contour_point = np.zeros([1, 2], dtype=np.uint8)
                contour_point[0][0] = j_contour
                contour_point[0][1] = i_contour
                # print(contour_point)
                if line.above != 0:
                    newProbAbove = int(line.above.bi_variate_gaussian_density(
                        contour_point))
                else:
                    newProbAbove = 0

                if line.below != 0:
                    newProbBelow = int(line.below.bi_variate_gaussian_density(
                        contour_point))
                else:
                    newProbBelow = 0

                addPrimesToList(newProbAbove, probAbovePrimes)
                addPrimesToList(newProbBelow, probBelowPrimes)
                # print(newProbAbove, newProbBelow)

        prob_above = 0
        prob_below = 0

        for k in range(len(probAbovePrimes)):
            mini = min(probAbovePrimes[k], probBelowPrimes[k])

            probAbovePrimes[k] -= mini
            probBelowPrimes[k] -= mini

            prob_above += probAbovePrimes[k] * primes[k]
            prob_below += probBelowPrimes[k] * primes[k]


        return prob_above < prob_below, line, contour
```

```python
def repair_lines(initial_lines, avg_line_height):
        """
        repeair all initial lines and generate the final line region
        """

        for line in initial_lines:
            column_processed = {}  # int, bool

            for column in range(img.shape[1]):#cols
                column_processed[column] = False

            i = 0
            while i < len(line.points):
                point = line.points[i]
                x = int(point[0])
                y = int(point[1])
                # print(y)
                # Check for vertical line intersection
                # In lines, we don't save all the vertical points we save only the start point and the end point.
                # So in line->points all we save is the horizontal points so, we know there exists a vertical line by
                # comparing the point[i].x (row) with point[i-1].x (row)
                if thresh_img[x][y] == 255:
                    if i == 0:
                        i+=1
                        continue
                    black_found = False
                    if line.points[i - 1][0] != line.points[i][0]:
                        # Means the points are in different rows (a vertical line).
                        min_row = int(min(line.points[i - 1][0], line.points[i][0]))
                        max_row = int(max(line.points[i - 1][0], line.points[i][0]))

                        for j in range(int(min_row), int(max_row) + 1):
                            if black_found is True:
                                break
                            if thresh_img[j][line.points[i - 1][1]] == 0:
                                x = j
                                y = line.points[i - 1][1]
                                black_found = True

                    if black_found == False:
                        i+=1
                        continue

                # Ignore it's previously processed

                if column_processed[y] == True:
                    i+=1
                    continue

                # Mark column as processed
                column_processed[y] = True

                avg_line_height = int(avg_line_height)

                for c in contours:
                    # Check line & contour intersection
                    tl = (c[0], c[1])
                    br = (c[0] + c[2], c[1] + c[3])

                    if y >= tl[0] and y <= br[0] and x >= tl[1] and x <= br[1]:
                        # print("br: {}".format(br))
                        # If contour is longer than the average height ignore
                        height = br[1] - tl[1]

                        if height > int(avg_line_height * 0.9):
                            continue

                        is_component_above, line, c = component_belongs_to_above_region(line, c)

                        # print(is_component_above)
                        new_row = 0
                        if is_component_above == False:
                            new_row = tl[1]
                            line.min_row_pos = min(line.min_row_pos, new_row)
                        else:
                            new_row = br[1]
                            line.max_row_pos = max(new_row, line.max_row_pos)

                        width = c[2]

                        for k in range(tl[0], tl[0] + width):
                            point_new = (new_row, line.points[k][1]) # make this coz tuple does not have value-assignment
                            line.points[k] = point_new

                        i = br[0]  # bottom right

                        # print("I: {}".format(i))
                        break  # Contour found
                i += 1

        return initial_lines
```

```python
initial_lines = repair_lines(initial_lines, avg_line_height)
```
